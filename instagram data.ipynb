```python
import string
string.punctuation
```

```python
def remove_punctuation(text):
    text_nopunt="".join([c for c in text if c not in string.punctuation])
    return text_nopunt
```

```python
data_1["text"]=data_1["text"].apply(lambda x:remove_punctuation(x))
```

```python
import nltk
nltk.download("stopword")
from nltk.corpus import stopwords
stop=stopwords.words("english")
stop
```

```python
data_1['text'] = data_1['text'].apply(lambda x: ' '.join([word for word in x.split() if word not in (stop)]))
```

```python
import re
nltk.download("words")
words=set(nltk.corpus.words.words())
words
```

```python
data_1.iloc[530,0]
```

```python
data_1.iloc[512,0]
```

```python
def clean_text_round2(text):
    '''Get read of some additional punctuation and non-sensical text that was missed the first time around.'''
    text = re.sub('government|politics|stategovernment|centralgovernment', '', text) 
    text = re.sub('protests|protest', '', text) 
    text = re.sub(r"\b[a-zA-Z]\b", "", text) 
    text = re.sub(r"b[a-zA-Z][a-zA-Z]\b", "", text)
    text =  " ".join(w for w in nltk.wordpunct_tokenize(text)  
                    if w.lower() in words)
    return text
round2 = lambda x: clean_text_round2(x)
```

```python
data_1 = pd.DataFrame(data_1.text.apply(round2))
data_1.iloc[530,0]
```

```python
 #Lemmatization
```

```python
from nltk.stem import WordNetLemmatizer 
import nltk
nltk.download('wordnet')
lemmatizer = WordNetLemmatizer()
w_tokenizer = nltk.tokenize.WhitespaceTokenizer()
```

```python
def lemmatize_text(text):
     return [lemmatizer.lemmatize(w,"v") for w in w_tokenizer.tokenize(text)]
```

```python
data_1.iloc[530,0]
```

```python
data_1['text'] = data_1.text.apply(lambda x :' '.join(lemmatize_text(x)))
data_1.iloc[530,0]
```

```python
from sklearn.feature_extraction.text import CountVectorizer

cv = CountVectorizer(stop_words='english')
data_cv = cv.fit_transform(data.text)
data_dtm = pd.DataFrame(data_cv.toarray(), columns=cv.get_feature_names())

tdm = data_dtm.transpose()
tdm.head()
```

```python
tdm['freq'] = tdm.sum(axis=1)
tdm.head(15)
```

```python
tdm.reset_index(inplace=True)
tdm.head()
```

```python
tdm1=tdm[["index","freq"]]
tdm1.rename(columns = {'index':'Word'}, inplace = True)
tdm1.sort_values(by='freq',ascending=False,inplace=True)
tdm1.head(20)
```

```python
import seaborn as sns
w = tdm1[tdm1['freq'] >=350] 
fg = sns.factorplot(x='Word', y='freq',
                        size=6,  aspect=2,
                        kind='bar', 
                        data=w)
```
